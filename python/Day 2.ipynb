```python
with open('day2_task1_input.txt', 'r') as f:
    password_input = f.readlines()
```

```python
def split_lines(line):
    occurance, key, password = line.split()
    key = key[0]
    password = password.strip()
    min_occurance = int(occurance.split('-')[0])
    max_occurance = int(occurance.split('-')[1])
    
    return {'key':key, 'password':password,
            'min_count':min_occurance, 'max_count':max_occurance}
```

```python
split_input = [split_lines(line) for line in password_input]
split_input
```

```python
def test_password(password_line):
    letters = ''
    for letter in password_line['password']:
        if letter == password_line['key']:
            letters += password_line['key']
    if (len(letters) >= password_line['min_count']) and (len(letters) <= password_line['max_count']):
        return 1
    else:
        return 0
```

```python
sum([test_password(line) for line in split_input])
```

```python
def test_password_corrected(password_line):
    if password_line['password'][password_line['min_count']-1] == password_line['key']:
        letter1 = 1
    else:
        letter1 = 0
        
    if password_line['password'][password_line['max_count']-1] == password_line['key']:
        letter2 = 1
    else:
        letter2 = 0
        
    if letter1 != letter2:
        return 1
    else:
        return 0
```

```python
sum([test_password_corrected(line) for line in split_input])
```
