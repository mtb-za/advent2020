```python
with open('day1_task1_input.txt', 'r') as f:
    nums = f.readlines()
```

```python
nums = [int(num) for num in nums]
nums
```

```python
pairs = [(x, y) for x in nums for y in nums]
```

```python
for pair in pairs:
    if sum(pair) == 2020:
        print(pair)
        print(pair[0] * pair[1])
```

```python
triples = [(x, y, z) for x in nums for y in nums for z in nums]
```

```python
for triple in triples:
    if sum(triple) == 2020:
        print(triple)
        print(triple[0] * triple[1] * triple[2])
```
