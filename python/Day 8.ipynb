```python
test_input = '''nop +0
acc +1
jmp +4
acc +3
jmp -3
acc -99
acc +1
jmp -4
acc +6'''.split('\n')

test_input = [ii.split() for idx, ii in enumerate(test_input)]
test_input
```

```python
def step(idx, instructions=[], accumulator=0, seen=[]):
    inst, val = instructions[idx]
    
    if idx in seen:
        return int(accumulator)
    seen.append(idx)
    
    if inst == 'nop':
        accumulator = step(idx + 1, instructions=instructions, accumulator=accumulator, seen=seen)
    if inst == 'acc':
        accumulator += int(val)
        accumulator = step(idx + 1, instructions=instructions, accumulator=accumulator, seen=seen)
    if inst == 'jmp':
        accumulator = step(idx + int(val), instructions=instructions, accumulator=accumulator, seen=seen)
    return accumulator
```

```python
def is_valid(idx, instructions=[], accumulator=0, seen=[]):
    inst, val = instructions[idx]
    
    if idx in seen:
        return False
    elif idx + 1 == len(instructions):
        return accumulator
    seen.append(idx)
    
    if inst == 'nop':
        valid = is_valid(idx + 1, instructions=instructions, accumulator=accumulator, seen=seen)
    if inst == 'acc':
        accumulator += int(val)
        valid = is_valid(idx + 1, instructions=instructions, accumulator=accumulator, seen=seen)
    if inst == 'jmp':
        valid = is_valid(idx + int(val), instructions=instructions, accumulator=accumulator, seen=seen)
        
    accumulator = valid
    
    if valid:
        return accumulator
    else:
        return False
```

```python
def excecute(instructions=[]):
    accumulator = 0
    idx = 0
    seen = []
    
    while True:
        if idx in seen:
            return False
        elif idx == len(instructions):
            break
        seen.append(idx)
            
        inst, val = instructions[idx]
        if inst == 'acc':
            accumulator += int(val)
            idx += 1
        elif inst == 'jmp':
            idx += int(val)
        elif inst == 'nop':
            idx += 1
        
    return accumulator
```

```python
def check_possibles(data):
    possibles = []
    swap = {'nop': 'jmp', 'jmp': 'nop'}
    for ii in range(len(data)):
        if data[ii][0] == 'nop' or data[ii][0] == 'jmp':
            possible = [[j for j in row] for row in data]
            possible[ii][0] = swap[possible[ii][0]]
            possibles.append(possible)
            result = is_valid(0, possible, accumulator=0, seen=[])
            if result is not False:
                return result
```
