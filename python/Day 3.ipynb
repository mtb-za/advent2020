```python
def count_hits(step_across, step_down=1, data=trees):
    trees_hit = []
    position = 0
    for ii, row in enumerate(data):
        if ii % step_down == 0:
            check = row[position % len(row)]
            trees_hit.append((check))#, position))
            position += step_across

    return trees_hit.count('#')
```

```python
def part1(data, r, d):
    n, m = len(data), len(data[0])
    j = 0
    res = 0
    for i in range(0, n, d):
        if data[i][j] == '#':
            res += 1
        j = (j+r) % m
    return res

part1(test_data, 2, 1)
```
